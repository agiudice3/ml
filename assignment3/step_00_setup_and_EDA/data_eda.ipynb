```python
import sys

from scipy.stats import skew, kurtosis
import pandas as pd
import numpy as np

sys.path.append("../")
from utils import load_data, StrokeParams, FetalHealthParams

RANDOM_STATE = 666
np.random.seed(RANDOM_STATE)

if __name__ == "__main__":
    # load data
    strokeX, strokeY = load_data(StrokeParams.data_train, StrokeParams.target_column)
    strokeX.drop(columns=StrokeParams.cat_cols, inplace=True)
    fetalX, fetalY = load_data(FetalHealthParams.data_train, FetalHealthParams.target_column)
    fetalX.drop(columns=FetalHealthParams.cat_cols, inplace=True)
```

```python
if __name__ == "__main__":
    # Calculate skewness and kurtosis for each feature
    skewness_stroke = strokeX.apply(skew, axis=0)
    kurtosis_stroke = strokeX.apply(kurtosis, axis=0, fisher=True)  # Fisher's definition, so normal kurtosis is 0

    skewness_fetal = fetalX.apply(skew, axis=0)
    kurtosis_fetal = fetalX.apply(kurtosis, axis=0, fisher=True)  # Fisher's definition

    # Combine results into a DataFrame
    skewness_kurtosis_stroke = pd.DataFrame({'Skewness': skewness_stroke, 'Kurtosis': kurtosis_stroke})
    skewness_kurtosis_fetal = pd.DataFrame({'Skewness': skewness_fetal, 'Kurtosis': kurtosis_fetal})

    display(skewness_kurtosis_stroke)
    display(skewness_kurtosis_fetal)
```
