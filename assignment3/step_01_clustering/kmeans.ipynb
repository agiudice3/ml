# Generate Clusters

```python
import sys

import pandas as pd
from sklearn.cluster import KMeans

sys.path.append("../")
from utils import load_data, StrokeParams, FetalHealthParams

RANDOM_STATE = 666
N_CLUSTERS = 2

if __name__ == "__main__":
    X_stroke_all, y_stroke = load_data(
        StrokeParams.data_train, StrokeParams.target_column
    )
    X_stroke = X_stroke_all.drop(columns=StrokeParams.cat_cols)

    X_stroke_test_all, y_stroke_test = load_data(
        StrokeParams.data_test, StrokeParams.target_column
    )
    X_stroke_test = X_stroke_test_all.drop(columns=StrokeParams.cat_cols)

    X_fetal_all, y_fetal = load_data(
        FetalHealthParams.data_train, FetalHealthParams.target_column
    )
    X_fetal = X_fetal_all.drop(columns=FetalHealthParams.cat_cols)

    X_fetal_test_all, y_fetal_test = load_data(
        FetalHealthParams.data_test, FetalHealthParams.target_column
    )
    X_fetal_test = X_fetal_test_all.drop(columns=FetalHealthParams.cat_cols)
```

```python
if __name__ == "__main__":
    # stroke
    km_stroke = KMeans(n_clusters=N_CLUSTERS, random_state=RANDOM_STATE, n_init=10)
    km_stroke.fit(X_stroke)

    X_stroke["km_cluster"] = km_stroke.predict(X_stroke)
    X_stroke_test["km_cluster"] = km_stroke.predict(X_stroke_test)

    df_train = pd.concat([X_stroke, y_stroke], axis=1)
    df_test = pd.concat([X_stroke_test, y_stroke_test], axis=1)

    df_train.to_csv(f"strokeX_kmeans_{N_CLUSTERS}.csv", index=False)
    df_test.to_csv(f"strokeX_test_kmeans_{N_CLUSTERS}.csv", index=False)

    # stroke all
    km_stroke_all = KMeans(n_clusters=N_CLUSTERS, random_state=RANDOM_STATE, n_init=10)
    km_stroke_all.fit(X_stroke_all)

    X_stroke_all["km_cluster"] = km_stroke_all.predict(X_stroke_all)
    X_stroke_test_all["km_cluster"] = km_stroke_all.predict(X_stroke_test_all)

    df_train_all = pd.concat([X_stroke_all, y_stroke], axis=1)
    df_test_all = pd.concat([X_stroke_test_all, y_stroke_test], axis=1)

    df_train_all.to_csv(f"strokeX_all_kmeans_{N_CLUSTERS}.csv", index=False)
    df_test_all.to_csv(f"strokeX_test_all_kmeans_{N_CLUSTERS}.csv", index=False)
```

```python
if __name__ == "__main__":
    # fetal health
    km_fetal = KMeans(n_clusters=N_CLUSTERS, random_state=RANDOM_STATE, n_init=10)
    km_fetal.fit(X_fetal)

    X_fetal["km_cluster"] = km_fetal.predict(X_fetal)
    X_fetal_test["km_cluster"] = km_fetal.predict(X_fetal_test)

    df_train = pd.concat([X_fetal, y_fetal], axis=1)
    df_test = pd.concat([X_fetal_test, y_fetal_test], axis=1)

    df_train.to_csv(f"fetalX_kmeans_{N_CLUSTERS}.csv", index=False)
    df_test.to_csv(f"fetalX_test_kmeans_{N_CLUSTERS}.csv", index=False)

    # fetal health all
    km_fetal_all = KMeans(n_clusters=N_CLUSTERS, random_state=RANDOM_STATE, n_init=10)
    km_fetal_all.fit(X_fetal_all)

    X_fetal_all["km_cluster"] = km_fetal_all.predict(X_fetal_all)
    X_fetal_test_all["km_cluster"] = km_fetal_all.predict(X_fetal_test_all)

    df_train_all = pd.concat([X_fetal_all, y_fetal], axis=1)
    df_test_all = pd.concat([X_fetal_test_all, y_fetal_test], axis=1)

    df_train_all.to_csv(f"fetalX_all_kmeans_{N_CLUSTERS}.csv", index=False)
    df_test_all.to_csv(f"fetalX_test_all_kmeans_{N_CLUSTERS}.csv", index=False)
```

```python
import sys

import pandas as pd
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

sys.path.append("../")
from utils import load_data, StrokeParams, FetalHealthParams
from step_01_clustering.plotting import (
    silhouette_plot,
    plot_tsne_clusters,
    plot_pca_results,
)

RANDOM_STATE = 666

datasets = [
    (
        "PCA",
        "fetal health",
        "../step_02_dim_red/fetal_health_pca_8.csv",
        "fetal_health",
        2,
    ),
    ("UMAP", "fetal health", "../step_02_dim_red/fetalX_umap_4.csv", "fetal_health", 2),
    (
        "PCA",
        "stroke prediction (all)",
        "../step_02_dim_red/stroke_all_pca_11.csv",
        "stroke",
        2,
    ),
    (
        "UMAP",
        "stroke prediction (all)",
        "../step_02_dim_red/strokeX_umap_13.csv",
        "stroke",
        2,
    ),
]

for algo, name, path, target_variable, n_components in datasets:
    df = pd.read_csv(path)
    X = df.drop(columns=target_variable)
    y = df[target_variable]

    silhouette_plot(
        X,
        n_components,
        algorithm="kmeans",
        dataset_name=f"{algo} {name.title()}",
    )

    tsne = TSNE(n_components=2, random_state=RANDOM_STATE)
    X_tsne = tsne.fit_transform(X)

    plot_tsne_clusters(
        X_original=X,
        X_tsne=X_tsne,
        true_labels=y,
        title=f"{algo} {name.title()} Dataset t-SNE Visualization",
        filename=f"{algo}_{target_variable}_tsne.png",
        n_clusters=2,
        n_components=2,
        random_state=RANDOM_STATE,
    )

    # pca = PCA(n_components=2, random_state=RANDOM_STATE)
    # X_pca = pca.fit_transform(X)

    # km = KMeans(
    #     n_clusters=2, random_state=RANDOM_STATE, n_init=10
    # )
    # km_labels = km.fit_predict(X_pca)
    # plot_pca_results(
    #     X_pca,
    #     y,
    #     km_labels,
    #     km,
    #     f"{algo} {name.title()} Dataset PCA",
    #     f"{algo}_{target_variable}_pca.png",
    # )
```
